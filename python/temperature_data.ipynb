### 자료출처
 - 지역별 수온온도 : 기상청(https://data.kma.go.kr/data/sea/selectBuoyRltmList.do?pgmNo=52)

```python
shinan_df=pd.read_csv('./data/temperature/신안 통합.csv',encoding='cp949',index_col='일시')
df1=shinan_df.iloc[:,1:].groupby('일시').mean()
df1['지점']=22183
df1.to_csv('./data/temperature/month/신안_월별.csv')
```

```python
geoje_df=pd.read_csv('./data/temperature/거문도 통합.csv',encoding='cp949',index_col='일시')
df2=geoje_df.iloc[:,1:].groupby('일시').mean()
df2['지점']=22103
df2.to_csv('./data/temperature/month/거문도_월별.csv')
```
