```python
shinan_df=pd.read_csv('./data/temperature/신안 통합.csv',encoding='cp949',index_col='일시')
shinan_df.head()
df1=shinan_df.iloc[:,1:].groupby('일시').mean()
df1['지점']=22183
df1.to_csv('./data/temperature/신안_월별.csv')
```
